```python
import json
import pandas as pd
from treelib import Node, Tree
from treelib.tree import NodeIDAbsentError
from treelib.tree import DuplicatedNodeIdError  
```

```python
src_fn="../../dataset/downstream_task/classification/dictionary.json"
```

```python
with open(src_fn,"r") as file:
    nodes=json.load(file)
```

```python
len(nodes)
```

```python
# Create tree object
tree = Tree()
```

```python
nodes_df=pd.read_json(src_fn).T
```

```python
nodes_df[nodes_df['concept_id']=="5709"]
```

```python
parent_nodes_df=nodes_df[nodes_df["parents"].apply(len)==0]
```

```python
non_parent_nodes_df=nodes_df[nodes_df["parents"].apply(len)!=0]
```

```python
non_parent_nodes_df=non_parent_nodes_df.explode("parents")
```

```python
tree=Tree()
tree.create_node(tag="root",identifier='-1')
```

```python
parent_nodes_df.apply(lambda row:tree.create_node(tag=row.label,identifier=row.concept_id,parent="-1"),axis=1)
```

```python
def create_node(row,count):
    try:
        tree.create_node(tag=row.label,identifier=row.concept_id,parent=row.parents)
    except (NodeIDAbsentError,DuplicatedNodeIdError):
        count["count"]+=1
        print(count["count"])
```

```python
count={"count":1}
non_parent_nodes_df.apply(lambda row:create_node(row,count),axis=1)
```

```python
print(tree.show())
```

```python
tree.save2file('tree.txt')
```
